## Part (a): 
Write a **Python Program** to find the value of unknown positive integer U using a series of questions. The question you can asked for a variable X are from these three Boolean answerable questions
1. `X>U`
2. `X<U`
3. `X==U`

The true value of question `X=U` signifies that you found the value of `U` as `X`. 
Range of U is not given and the value of U can be any positive integer (may be very big number)

You are supposed to come up with an approach to find the value of U with asking minimum number of questions. Your program should output the number of quesion asked (or number of times you performed comparison with U).

**Sample input for U:**
1. 50
2. 1332060
3. 6576
4. 1800

---

### Approach:

Step1: First find the upper bound using an exponential function, calculate the comparison along with the loop

Step2: Then apply binary search, calculate the comparison along with the loop

Step3: Return the no. of comparison at last.


```python
def fun(key):


    def find_upper_bound(end,key,comparison):
        while(key>end):
            comparison+=1
            # print(f"Start: {start}, End: {end}")
            end = end*2
        return end, comparison
    

    def binary_search(start,end,key,comparison):
        while(start<=end):
            comparison+=1
            mid = start + (end-start)//2
            if(mid==key):
                # print(f"Start: {start}, End: {end}")
                return comparison
            elif(mid>key):
                # print(f"Start: {start}, End: {end}")
                end = mid-1
            else:
                # print(f"Start: {start}, End: {end}")
                start = mid+1

    
    start = 1
    end = 1
    comparison = 0
    new_end, new_comparison = find_upper_bound(end,key,comparison)
    comparison = binary_search(start, new_end, key, new_comparison)

    return comparison

a = 50
b = 1332060
c= 6576
d = 1800
fun(a)
```

```python
fun(d)
```

## Part (b):
Write a **Python Program** to find the square root of a positive integer up to accuracy of 5th decimal place. You are not supposed to use the inbuilt square root function to calculate the square root of the number

**Sample input:**
1. 50
2. 2000
3. 6576
4. 1800

Approach 
1. Brute: We can traverse lineraly, from 1-->Num, but it will be solved in O(N)
2. Optimal: Tranverse using binary search. TC=O(logN)

```python
def square_root(N):
    if N == 0 or N == 1:
        return float(N)

    low = 0.0
    high = float(N)
    precision = 1e-6   # for 5 decimal places

    while high - low > precision:
        mid = (low + high) / 2

        if mid * mid < N:
            low = mid
        else:
            high = mid

    return round((low + high) / 2, 5)
```

```python
square_root(50)
```

```python
square_root(2000)
```

```python
square_root(6576)
```

```python
square_root(1800)
```
